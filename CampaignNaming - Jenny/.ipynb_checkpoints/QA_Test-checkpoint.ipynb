```python
import pandas as pd
import numpy as np
from pandas import Series
from pandas import read_csv
from openpyxl import load_workbook
from collections import defaultdict
```

```python
path = "/Users/zhangwang/Desktop/EDL/QA Project/"
file = "Non-Paid Social Campaign Naming QA Problem - 2019-05-29 - pm .xlsx"
rules = "Rules.xlsx"

campaign = pd.ExcelFile(path+file).parse("Test Google Ads Campaigns")
```

```python
rule = pd.ExcelFile(path+rules)
rlst = []
for i in range(len(rule.sheet_names)-2):
    rlst.append(rule.sheet_names[i])

rlst
#get the sheet names for the rules
```

```python
#get rules for different scenarios from each page as a list 

UDS = list(rule.parse(str(rlst[0]))["UDS"])
EDL = list(rule.parse(str(rlst[1]))["EDL"])
CLNT = list(rule.parse(str(rlst[2]))["CLNT"])
PRD = list(rule.parse(str(rlst[3]))["PRD"])
#CHNL = list(rule.parse(str(rlst[4]))["CHNL"])
#TACT = list(rule.parse(str(rlst[5]))["TACT"])
BNM = list(rule.parse(str(rlst[5]))["BNM"])
CRI = list(rule.parse(str(rlst[6]))["CRI"])
NLT = list(rule.parse(str(rlst[7]))["NLT"])
CNT = list(rule.parse(str(rlst[8]))["CNT"])
LNG = list(rule.parse(str(rlst[9]))["LNG"])
ENG = list(rule.parse(str(rlst[10]))["ENG"])

#match channel and Tactics, then combined into a dictionary
test = rule.parse("CHNL_TACT")
CHNL_TACT = defaultdict(list)

for index, row in test.iterrows():
    CHNL_TACT[row["CHNL"]].append(row["TACT"])

#print(CHNL_TACT)
```

```python
QA1 = []
QA2 = []
QA3 = []
QA4 = []
QA5 = []
QA6 = []
QA7 = []
QA8 = []
QA9 = []
QA10 = []
QA11 = [] 
result = []  
```

## QA1: 10 Underscores Check
**Check whether there are 10 underscores for each campaign name**

```python
def QA1_test(lst):
    if len(lst) == UDS[0]:
        QA1.append(0)
    else:
        QA1.append(1)
        
    return QA1
```

## QA2: EDL Check
**Check whether the client is for EDL vs. zzzEDL**

```python
def QA2_test(a, lst):
    if lst[a] in EDL:
        QA2.append(0)
    else:
        QA2.append(1)
        
    return QA2
```

## QA3: CNLT Check
**Check whether the client name is listed in the Client Short Name List**

```python
def QA3_test(a, lst):
    if lst[a] in CLNT:
        QA3.append(0)
    else:
        QA3.append(1)
        
    return QA3
```

## QA4: PRD Check
**Check whether the # of "-" in product is less and equal to 5**

```python
def QA4_test(a, lst):
    if len(lst[a].split("-"))<=PRD[0] or "-" not in lst[a]:
        QA4.append(0)
    else:
        QA4.append(1)
        
    #return plst
    return QA4
```

## QA5: CHNL_TACT Check
**Check whether Channel and Tactic match with each other, and the campaign not Social**

```python
def QA5_test(a, lst):
    if lst[a] in CHNL_TACT and lst[a+1] in CHNL_TACT.get(lst[a]):
        QA5.append(0)
    else: 
        QA5.append(1)

    return QA5
```

## QA6: BNM Check
**Check Brand/Non/Comp**

```python
def QA6_test(a, lst):
    if lst[a] in BNM:
        QA6.append(0)
    else:
        QA6.append(1)
    
    return QA6
```

## QA7: CRI Check
**Check Criterion Type**

```python
def QA7_test(a, lst):
    if lst[a] in CRI:
        QA7.append(0)
    else:
        QA7.append(1)

    return QA7
```

## QA8: NLT Check
**Check NLT length, meaning the # of dashes < = 5**

```python
def QA8_test(a, lst):
    if len(lst[a].split("-"))<=NLT[0] or "-" not in lst[a]:
        QA8.append(0)
    else:
        QA8.append(1)

    return QA8
```

## QA9: CNT Check
**Check whether the campaign is in the target country**

```python
def QA9_test(a, lst):
    if lst[a] in CNT:
        QA9.append(0)
    else:
        QA9.append(1)
        
    return QA9
```

## QA10: LNG Check
**Check whether the campaign is set in the target language**

```python
def QA10_test(a, lst):
    if lst[a] in LNG:
        QA10.append(0)
    else:
        QA10.append(1)
        
    return QA10
```

## QA11: ENG Check
**Check whether the campaign is set in the target engine**

```python
def QA11_test(a, lst):
    if lst[a] in ENG:
        QA11.append(0)
    else:
        QA11.append(1)

    return QA11
```

## QA Check for the whole campaign name file

```python
def QA_check():
    for i in range(len(campaign["Campaign name"])):
        lst = campaign["Campaign name"][i].split("_")
        QA1_test(lst)
        a = 0
        QA2_test(a,lst)
        a += 1
        QA3_test(a,lst)
        a += 1
        QA4_test(a,lst)
        a += 1
        QA5_test(a,lst)
        a += 2
        QA6_test(a,lst)
        a += 1
        QA7_test(a,lst)
        a += 1
        QA8_test(a,lst)
        a += 1
        QA9_test(a,lst)
        a += 1
        QA10_test(a,lst)
        a += 1
        QA11_test(a,lst)   
        
        QA = QA1[i]+QA2[i]+QA3[i]+QA4[i]+QA5[i]+QA6[i]+QA7[i]+QA8[i]+QA9[i]+QA10[i]+QA11[i]
        if QA == 0:
            result.append("Pass")
        else:
            result.append("Fail")

QA_check()
```

```python
#combine QA Check result into a 
name = list(campaign["Campaign name"])
QA_result = pd.DataFrame(list(zip(name, result, QA1, QA2, QA3, QA4, QA5, QA6, QA7, QA8, QA9, QA10, QA11)),
                     columns = ["Campaign name","Pass/Fail","Underscores","EDL","Client Short Name","Product",
                               "Channel & Tactic","Brand/Non/Comp","Criterion Type","NLT","Country","Language",
                               "Engine"])
```
